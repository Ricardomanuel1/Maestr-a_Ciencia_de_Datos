Cálculo de probabilidad de número de hijos (k)

```python
n=10
```

```python
for k in range(1,n+1):
  ro=k/(2**(k+1))
  print(f'La probabilidad de tener {k} hijos es {ro}')
```

```python
k=60
k_values = np.arange(1, k+1)
def ro(k):
    return k / (2**(k+1))
plt.figure(figsize=(10, 6))
plt.bar(k_values, ro(k_values), width=0.8, align='center', alpha=0.7)
plt.title('Distribución de p(k) = k / (2^(k+1)) k>=1')
plt.xlabel('k')
plt.ylabel('ro(k)')
plt.xticks(k_values,rotation=90)
plt.show()
```

```python
#Observamos que la probabilidad a partir del 7 hijo ba disminuyendo siendo cercana a cero.
```

Acumulación de probabilidades

```python
n=10
```

```python
#Intervalos creados a partir de la probabilidad de tener n hijos
lista=[]
c=0
for k in range(1,n+1):
  ro=k/(2**(k+1))
  c=c+ro
  lista.append(c)
lista
```

```python
#Generamos un número aleatorio uniforme entre cero y uno
aleatorio = np.random.uniform(0, 1)
aleatorio
```

```python
#Según en que intervalo de la suma de probabilidades se encuentra el número aleatorio uniforme hallamos el número de hijos
def estado(aleatorio, lista):
     #acumula=[]
     for i in range(len(lista)):
        if aleatorio < lista[i]:
          #acumula.append(i+1)
          return i + 1
          #return acumula

print('El número de hijos es:',estado(aleatorio,lista))
```
