```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Experimento 1: Lanzamiento de una moneda

```python
# Experimento 1: Lanzar una moneda
p = 0.5 # probabilidad de obtener cara
nsim = 10000 # num de experimentos
nhead = 0 # num de caras obtenidas
salida = [] # armazena as saidas (cara:1, sello:0)
for i in range(0, nsim):
    if(np.random.uniform() < p): # se es menor que p, cara
        nhead = nhead + 1 # incrementa el contador de caras
        salida.append(1)
    else:
        salida.append(0)
print("Salida:", salida)
print("Frequencia de caras:", nhead/nsim)
```

```python
p = 0.5
vp = [] # frequencia de ocurrencias
vsim = [] # número de simulaciones
Nmax = 1000 # número máximo de simulaciones
for nsim in np.arange(1,Nmax+1,10):
    nhead = 0 # numero de caras
    for i in range(1,nsim):
        if(np.random.uniform() < p):
            nhead = nhead + 1
    vp.append(nhead/nsim)
    vsim.append(nsim)
# Mostra os resultados em uma figura
plt.figure(figsize=(25,4))
plt.plot(vsim, vp, linestyle='-', color="blue", linewidth=2,label = 'Valor Estimado')
plt.axhline(y=p, color='r', linestyle='--', label = 'Valor Teórico')
plt.ylabel("Probabilidad", fontsize=12)
plt.xlabel("Número de Simulaciones", fontsize=12)
plt.xlim([0.0, Nmax+1])
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.show(True)
```

# Experimento 3: El problema de Monty Hall

```python
import random
from random import seed
seed(100)

puertas = [1,2,3]
Nmax = 1000
P = [] # Guarda la fracción de victorias
vN = [] # Guarda el número de juegos
for N in np.arange(1, Nmax,10):
    victoria = 0
    derrota = 0
    for i in range(0,N):
        C = random.choice(puertas) # Coloca el carro en una puerta
        X = random.choice(puertas) # Selecciona una puerta
        if(C != X): # Si el carro no está en la puerta seleccionada
            # el presendador abrirá otra porta que tiene a la cabra.
            # Cambiando de puerta, se gana el automóvil.
            victoria = victoria + 1
    P.append(victoria/N)
    vN.append(N)
plt.figure(figsize=(25,4))
plt.plot(vN, P, '-', color='blue', linewidth=2,label='Valor Estimado')
plt.axhline(y=2/3,color='red', linestyle='--', label='Valor teórico')
plt.ylabel("Probabilidad", fontsize=12)
plt.xlabel("Número de Simulaciones", fontsize=12)
plt.xlim([0.0, Nmax])
plt.ylim([0, 1.0])
plt.legend(loc='lower right')
plt.show(True)
```

# Experimento 4: El problema del Apostador

```python
from random import seed
from random import randint
# Semilla del gerador de números aleatórios
seed(1)

apuesta = 5
Ng = 1000 # Número de juegos
p = 0
for n in range(1,Ng):
    G = 0
    j = randint(1, 6) # Punto del Jugador
    b = randint(1, 6) # Punto del Oponente
    if(j > b): # Si el punto del Jugador es mayor
        G = 2*(j-b) # Ganancia en una jugada
    if(G > apuesta): # En caso haya Ganancia
        p = p + 1
p = p/Ng # Probabilidad de tener Ganancia
print('Apuesta:', apuesta, 'soles.')
print("Probabilidad de Ganar:", p)
print("Valor teórico:", 1/6)
```

```python
seed(100)

plt.figure(figsize=(25,4))
vp = [] # Vector que guarda el valor de la apuesta
vapuesta = [] # Vector que guarda la probabilidad de ganancia
for apuesta in np.arange(0,11,2):
    Ng = 100 # Número de juegos
    p = 0
    for n in range(1,Ng):
        G = 0 # Ganancia
        j = randint(1, 6) # Punto del Jugador
        b = randint(1, 6) # Punto del Oponente
        if(j > b): # Si el punto del Jugador es mayor
            G = 2*(j-b) # Ganancia en una jugada
        if(G > apuesta): # En caso haya Ganancia
            p = p + 1
    p = p/Ng # Probabilidad de tener Ganancia
    vp.append(p)
    vapuesta.append(apuesta)
plt.plot(vapuesta, vp, 'bo-', label = 'Valor Estimado')
plt.xlabel('Valor de la apuesta', fontsize = 12)
plt.ylabel('Probabilidad de Ganar',fontsize = 12)

# valor teorico
vpt = [15/36,10/26,6/36,3/36,1/36,0] #Probabilidad teórica de tener Ganancia
vapuestat=[0,2,4,6,8,10] # Valor de la Apuesta
plt.plot(vapuestat, vpt, 'ro-',label='Valor Teórico')
plt.legend(loc='upper right')
plt.show(True)
```

# Experimento 5: Estimación de $\pi$

```python
X=[]
Y=[]
n = 1000
count_inside = 0
for i in np.arange(0,n):
    x = np.random.uniform(0,1)
    y = np.random.uniform(0,1)
    X.append(x)
    Y.append(y)
    if( x**2 + y**2 < 1):
        count_inside = count_inside + 1
pi = 4*count_inside/n
print("Número de simulaciones:", n)
print("Número de puntos internos:", count_inside)
print("Valor de pi estimado:", pi)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(X,Y)
draw_circle = plt.Circle((0.5, 0.5), 0.5,fill=False,color='red',lw=2)
plt.gcf().gca().add_artist(draw_circle)
plt.show()
```
