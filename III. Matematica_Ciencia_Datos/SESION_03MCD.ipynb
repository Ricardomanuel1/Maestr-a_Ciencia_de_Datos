**Ejemplo:** Traslaciones

$$T_a : \mathbb R^2 \rightarrow  \mathbb R^2 \\ T_a(x,y)=(x+h,y+k)\\   a =(h,k) $$

```
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-2, 2,0.01)
y=np.sin(3*x**2)
from numpy import matrix
M=matrix([[1,0,8],[0,1,2],[0,0,1]])
import numpy as np
xx=np.ones(len(x))
X=matrix([x,y,xx])
AX=np.dot(M,X)
x1=np.ravel(AX[0],order='F')
x2=np.ravel(AX[1],order='F')
plt.plot(x,y,"r",x1,x2,"g")
plt.legend(['funcion f', 'Traslacion de f'])
```

## **Grafica con traslación**

```
plt.plot(x1,x2,"b")
```

## **Traslación inversa**

$T(x,y)=(x,y)+(h,k)$

$T'(x,y)=(x,y)+(-h,-k)$

```
h=-8
k=-2
M=matrix([[1,0,h],[0,1,k],[0,0,1]])
```

```
X=matrix([x1,x2,xx])
AX=np.dot(M,X)
x1=np.ravel(AX[0],order='F')
x2=np.ravel(AX[1],order='F')
```

```
plt.plot(x1,x2,"b")
```

**Ejemplo:** Rotación

$$T_{\theta} : \mathbb R^2 \rightarrow  \mathbb R^2 \\ T_{\theta}(x,y)=(xcos{\theta}-y{sen\theta},xsen{\theta}+y{cos\theta}),\theta \in \mathbb R $$

$$A=
\begin{bmatrix}
cos\theta & -sen\theta\\
sen\theta & cos\theta
\end{bmatrix}
$$


## **Rotación**

```
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-2, 2,0.01)
y=x**2
t=np.pi/2
from numpy import matrix
M=matrix([[np.cos(t),-np.sin(t),0],[np.sin(t),np.cos(t),0],[0,0,1]])
xx=np.ones(len(x))
X=matrix([x,y,xx])
AX=np.dot(M,X)
x1=np.ravel(AX[0],order='F')
x2=np.ravel(AX[1],order='F')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axis('square')
plt.plot(x,y,"r",x1,x2,"g")
plt.legend(['funcion f', 'Rotación de f'])
```

## **Grafica rotada**

```
plt.plot(x1,x2,"g")
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axis('square');
```

## **Matriz de rotación inversa**
$$$$

$$[T]_{\theta}=
\begin{bmatrix}
cos\theta & -sen\theta\\
sen\theta & cos\theta
\end{bmatrix}
\Longrightarrow
[T]^{-1}_{\theta}=
\begin{bmatrix}
cos\theta & sen\theta\\
-sen\theta & cos\theta
\end{bmatrix}
$$

```
M=matrix([[np.cos(t),np.sin(t),0],[-np.sin(t),np.cos(t),0],[0,0,1]])
```

```
X=matrix([x1,x2,xx])
AX=np.dot(M,X)
x1=np.ravel(AX[0],order='F')
x2=np.ravel(AX[1],order='F')
```

```
plt.plot(x1,x2,"b")
```

## **Usando un slider**

```
import numpy as np
import matplotlib.pyplot as plt
from IPython.html.widgets import interact
import ipywidgets as widgets
from numpy import matrix
```

```
def rot2(t):
    x = np.arange(-2, 2,0.01)
    y=x**2
    M=matrix([[np.cos(t),-np.sin(t),0],[np.sin(t),np.cos(t),0],[0,0,1]])
    xx=np.ones(len(x))
    X=matrix([x,y,xx])
    AX=np.dot(M,X)
    x1=np.ravel(AX[0],order='F')
    x2=np.ravel(AX[1],order='F')
    plt.xlim(-4, 4)
    plt.ylim(-4, 4)
    plt.axis('square')
    plt.plot(x,y,"r",x1,x2,"g")
```

```
interact(rot2,t=(0,np.pi/2));
```
