```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Ejemplo 1:
Una urna contiene cinco bolas numeradas 1, 2, 2, 8 y 8. Si
una persona selecciona tres bolas al azar y definimos la variable aleatoria $X$ como la suma obtenida por las tres bolas.



## (a) Calculando la Funci√≥n de Probabilidad - f(x)=Prob(X=x)

```python
#1,2,2,8,8
X=['Suma=5','Suma=11','Suma=12','Suma=17','Suma=18']
P=[0.1, 0.4,0.2,0.1,0.2]
```

```python
df=pd.DataFrame()
df['x']=X
df['f(x)']=P
df.set_index('x',inplace=True)
df
```

```python
plt.plot(df['f(x)'],marker='*')
plt.ylim(0,1.1)
plt.show()
```

## (b) Calculando la Funci√≥n de Probabilidad Acumulada - F(x)


```python
X=['Suma=5','Suma=11','Suma=12','Suma=17','Suma=18']
P=[0.1, 0.4,0.2,0.1,0.2]
```

```python
df=pd.DataFrame()
df['x']=X
df['f(x)']=P
df['F(x)']=np.cumsum(df['f(x)'])
df.set_index('x',inplace=True)
df
```

```python
plt.plot(df['F(x)'],marker='*')
plt.ylim(0,1.1)
plt.show()
```

## (c) Calcular E(X)

```python
df=pd.DataFrame()
X_ind=['Suma=5','Suma=11','Suma=12','Suma=17','Suma=18']
X=[5,11,12,17,18]
P=[0.1, 0.4,0.2,0.1,0.2]
df['x_ind']=X_ind
df['x']=X
df['f(x)']=P
df['x*f(x)']=df['x']*df['f(x)']
df.set_index('x_ind',inplace=True)
df
```

```python
Valor_Esperado=df['x*f(x)'].sum()
print('El Valor Esperado es', Valor_Esperado)
```

## (d) Calculando Varianza y Desviaci√≥n Estandar usando Formula 1
$$Var(X)= E([X-E(X)]^2)\quad\text{y}\quad sd(X)=\sqrt{Var(X)}.$$

```python
df=pd.DataFrame()
X_ind=['Suma=5','Suma=11','Suma=12','Suma=17','Suma=18']
X=[5,11,12,17,18]
P=[0.1, 0.4,0.2,0.1,0.2]
df['x_ind']=X_ind
df['x']=X
df['f(x)']=P
df['x*f(x)']=df['x']*df['f(x)']
Valor_Esperado=df['x*f(x)'].sum()
print('------------------------------------------------------------------------')
print('Tabla de Distribuci√≥n de Probabilidades')
print('------------------------------------------------------------------------')
df['x-E(X)']=df['x']-Valor_Esperado
df['[x-E(X)]^2']=df['x-E(X)']*df['x-E(X)']
df['[x-E(X)]^2 * f(x)']=df['[x-E(X)]^2']*df['f(x)']
df.set_index('x_ind',inplace=True)
Varianza=df['[x-E(X)]^2 * f(x)'].sum()
print(df)
print('------------------------------------------------------------------------')
print('E(X)=', Valor_Esperado)
print('------------------------------------------------------------------------')
print('Var(X)=',Varianza)
print('------------------------------------------------------------------------')
print('sd(X)',round(np.sqrt(Varianza),2))
print('------------------------------------------------------------------------')
```

## (e) Calculando Varianza y Desviaci√≥n Estandar usando Formula 2
$$Var(X)= E(X^2)-E(X)^2\quad\text{y}\quad sd(X)=\sqrt{Var(X)}.$$

```python
df=pd.DataFrame()
X_ind=['Suma=5','Suma=11','Suma=12','Suma=17','Suma=18']
X=[5,11,12,17,18]
P=[0.1, 0.4,0.2,0.1,0.2]
df['x_ind']=X_ind
df['x']=X
df['f(x)']=P
df['x*f(x)']=df['x']*df['f(x)']
Valor_Esperado=df['x*f(x)'].sum()
df['x^2']=df['x']*df['x']
df['x^2*f(x)']=df['x^2']*df['f(x)']
df.set_index('x_ind',inplace=True)
Valor_Esperado=df['x*f(x)'].sum()
Varianza=df['x^2*f(x)'].sum()-Valor_Esperado**2
```

```python
print('------------------------------------------------------------------------')
print('Tabla de Distribuci√≥n de Probabilidades')
print('------------------------------------------------------------------------')
print(df)
print('------------------------------------------------------------------------')
print('E(X)=', Valor_Esperado)
print('------------------------------------------------------------------------')
print('Var(X)=',round(Varianza,2))
print('------------------------------------------------------------------------')
print('sd(X)',round(np.sqrt(Varianza),2))
print('------------------------------------------------------------------------')
```

# Ejemplo 2:

## (a) Calculando la Funci√≥n de Probabilidad - f(x)

```python
#Efectuando el arbol de probabilidades
#X=0 GG    ...........(7/12)*(6/11)                 P(x=0)=0.318
#X=1 GA,AG ...........(7/12)*(5/11)+ (5/12)*(7/11)  P(x=1)=0.530
#X=2 AA    ...........(5/11)*(4/11)                 P(x=2)=0.165
X=['GG','GA-AG','AA']
P=[0.318, 0.53,0.165]
```

$$Var(X)= E([X-E(X)]^2)\quad\text{y}\quad sd(X)=\sqrt{Var(X)}.$$

```python
df=pd.DataFrame()
X_ind=['GG','GA-AG','AA']
X=[0,1,2]
P=[0.318,0.530,0.165]
df['x_ind']=X_ind
df['x']=X
df['f(x)']=P
df['x*f(x)']=df['x']*df['f(x)']
Valor_Esperado=df['x*f(x)'].sum()
print('------------------------------------------------------------------------')
print('Tabla de Distribuci√≥n de Probabilidades')
print('------------------------------------------------------------------------')
df['x-E(X)']=df['x']-Valor_Esperado
df['[x-E(X)]^2']=df['x-E(X)']*df['x-E(X)']
df['[x-E(X)]^2 * f(x)']=df['[x-E(X)]^2']*df['f(x)']
df.set_index('x_ind',inplace=True)
Varianza=df['[x-E(X)]^2 * f(x)'].sum()
print(df)
print('------------------------------------------------------------------------')
print('E(X)=', Valor_Esperado)
print('------------------------------------------------------------------------')
print('Var(X)=',Varianza)
print('------------------------------------------------------------------------')
print('sd(X)',round(np.sqrt(Varianza),2))
print('------------------------------------------------------------------------')
```

```python
df=pd.DataFrame()
X_ind=['GG','GA-AG','AA']
X=[0,1,2]
P=[0.318,0.530,0.165]
df['x_ind']=X_ind
df['x']=X
df['f(x)']=P
df['x*f(x)']=df['x']*df['f(x)']
Valor_Esperado=df['x*f(x)'].sum()
df['x^2']=df['x']*df['x']
df['x^2*f(x)']=df['x^2']*df['f(x)']
df.set_index('x_ind',inplace=True)
Valor_Esperado=df['x*f(x)'].sum()
Varianza=df['x^2*f(x)'].sum()-Valor_Esperado**2
```
