**For reading an html from url**

```
from bs4 import BeautifulSoup # Clean data from html format
#BeautifulSoup es una biblioteca de Python que se utiliza para analizar documentos HTML y XML

import urllib.request # Request URL
#urllib.request es un módulo en Python que se utiliza para abrir y leer URLs

response = urllib.request.urlopen('https://sites.google.com/site/erickgomeznieto/selected-publications')
html = response.read()

print(html)

# Then, we use BeautifulSoup to clean from tags

soup = BeautifulSoup(html,"html.parser")#se utiliza para limpiar con el parser de html
#permite analizar y manipular el HTML descargado

text = soup.get_text(strip=True)
#se utiliza para extraer todo el texto de un documento HTML, eliminando las etiquetas HTML
#strip=True se eliminan también los espacios en blanco adicionales alrededor del texto

original_tokens = [t for t in text.split()]#se crea una lista de palabras tokenizadas basado en el espacio en blanco

print(original_tokens)#lista de tokens
```

***Tokenize process***

```
from nltk.tokenize import word_tokenize
# generate tokens
#original_tokens = word_tokenize(text)
#print(original_tokens)
# to lower case
print(original_tokens)

original_tokens = [x.lower() for x in original_tokens]

print(original_tokens)
```

```
# Count frequencies
freq = nltk.FreqDist(original_tokens)

#for key,val in freq.items():
    # print(str(key) + ':' + str(val))

freq.plot(30,cumulative=False)
plt.show()
```

```
print(len(original_tokens))
```

***Cleaning Tokens 1 - Remove punctuation***

```
# remove all tokens that are not alphabetic, remueve los token no alfabeticos
#en el ejempo de los 30 token mas frecuente se borra el g.
clean1_tokens = [word for word in original_tokens if word.isalpha()]
print(len(clean1_tokens))
```

```
# Count frequencies
freq2 = nltk.FreqDist(clean1_tokens)

freq2.plot(30,cumulative=False)
plt.show()
```

***Cleaning Tokens 2 - Remove stopwords***

```
# Cleaning tokens
#clean2_tokens = clean1_tokens[:]#se hace una copia

# get all stopwords
sr = nltk.corpus.stopwords.words('english')#se carga la lista de los stopword en ingles
#sr.append("chicken")#si se desea agregar una nueva palabra
#print(sr)
#print(len(sr))

clean2_tokens = [word for word in clean1_tokens if not word in sr]

print(len(original_tokens))      #Los token en mayuscula se cambian a minuscula
print(len(clean1_tokens))        #Limpiando los no alfabeticos
print(len(clean2_tokens))        #Limpiando los stopword, los no alfabeticos

print(original_tokens)           #Los token en mayuscula se cambian a minuscula
print(clean2_tokens)             #limpiando los stopword, los no alfabeticos
```

```
# Count frequencies
freq3 = nltk.FreqDist(clean2_tokens)

freq3.plot(30,cumulative=False)
plt.show()
```

***Lemmatization***

```
# Lemmatization, en base a las palabras del diccionario

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_tokens = []

for token in clean2_tokens:
    lemmatized_tokens.append(lemmatizer.lemmatize(token))

print(clean2_tokens)
print(lemmatized_tokens)
```

```
# Calculate frequencies again!
freq_lemma = nltk.FreqDist(lemmatized_tokens)
#print(len(original_tokens))
#print(len(clean1_tokens))
#print(len(clean2_tokens))
#print(len(freq4.items()))
freq_lemma.plot(30,cumulative=False)
plt.show()
```

```
words = ['running', 'ate', 'better', 'geese','things']

lemmas = [lemmatizer.lemmatize(word) for word in words]

print("Lemmas en inglés:", lemmas)
```

# **Stemming**

```
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")

stemming_tokens = []

for token in clean2_tokens:
    stemming_tokens.append(SnowballStemmer("english").stem(token))

print(stemming_tokens)
```

```
# Calculate frequencies again!
freq_stem = nltk.FreqDist(stemming_tokens)
#print(len(original_tokens))
#print(len(clean1_tokens))
#print(len(clean2_tokens))
#print(len(freq4.items()))
freq_stem.plot(30,cumulative=False)
plt.show()
```

## Setting font size max and min value
