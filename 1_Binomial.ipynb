# Stochastic Simulation: Binomial Distribution

Simulemos el lanzamiento de $n$ monedas y la probabilidad de obtener $k$ caras. Osea, tenemos $k$ Ã©xitos en $n$ realizaciones de un experimento.

```python
from random import seed
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import binom

import math

seed(100) # Semilla del generador de nÃºmeros aleatorios

n = 100 # NÃºmero de lanzamientos
p = 0.3 # Probabilidad de obtener cara.
Pk = np.zeros(n)
vk = np.arange(0,n)
ns = 10000 # NÃºmero de simulaciones

for j in range(0,ns):
    S = 0 # NÃºmero de Ã©xitos
    for i in range(0,n):
        r = np.random.uniform() #
        if(r <= p): # Si existe Ã©xito
            S = S + 1
    Pk[S] = Pk[S] + 1
Pk=Pk/sum(Pk) # Normaliza la distribuciÃ³n
plt.figure(figsize=(10,6))
plt.xlim(0.8*np.min(vk[Pk>0]),1.2*np.max(vk[Pk>0]))
plt.bar(vk, Pk, label='SimulaciÃ³n')

# curva teÃ³rica
Pkt = np.zeros(n+1) # valores teÃ³ricos
vkt = np.arange(0,n+1)
for k in range(0,n+1):
    Pkt[k] = (math.factorial(n)/(math.factorial(n-k)*math.factorial(k)))*(p**k)*(1-p)**(n-k)
plt.plot(vkt, Pkt, 'r--', label='Prob. TeÃ³rica')

plt.xlabel('k', fontsize = 15)
plt.ylabel('P(k)',fontsize = 15)
plt.legend()
plt.show(True)
```
