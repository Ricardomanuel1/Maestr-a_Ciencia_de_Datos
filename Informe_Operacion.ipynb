```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
import zipfile
import plotly.express as px
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
import time
from matplotlib import style
import openpyxl
```

```python

ruta_zip='/content/OneDrive_2023-08-11.zip'
```

```python
# Abrir el archivo ZIP en modo lectura
with zipfile.ZipFile(ruta_zip, "r") as archivo_zip:
    # Extraer todos los archivos del ZIP
    archivo_zip.extractall()




```

```python
# Listar los archivos contenidos en el ZIP - almacenamos lista de archivos en una variable
lista_archivos = archivo_zip.namelist()
print("Archivos contenidos en el ZIP:", lista_archivos)
```

```python
lista_archivos
```

```python
# Guardar los enlaces en un archivo CSV
with open("enlaces.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Enlace"])
    writer.writerows([[enlace] for enlace in lista_archivos])

print("Enlaces exportados correctamente a enlaces.csv.")
```

```python
# Leer los enlaces desde el archivo CSV
enlaces_df = pd.read_csv("enlaces.csv")
```

```python
# Crear una lista para almacenar los DataFrames de cada enlace
dfs = []

# Iterar sobre los enlaces y leer cada archivo CSV
for enlace in enlaces_df["Enlace"]:
    df = pd.read_csv(enlace, sep=';', encoding='unicode_escape')
    dfs.append(df)

# Concatenar los DataFrames en uno solo
df_final = pd.concat(dfs)

# Realizar cualquier procesamiento adicional con el DataFrame df_final

# Mostrar el DataFrame final
df_final
```
