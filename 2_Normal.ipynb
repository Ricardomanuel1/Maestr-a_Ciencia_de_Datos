## Ejemplo 1:

Si $X \sim  \mathcal{N}(\mu=165,\,\sigma^2=9)$, calcule $P(X<162)$.

```python
mu = 165
sigma_2 = 9
sd = np.sqrt(sigma_2)
z = (162-mu)/sd
# P(X<162) ---> P(Z<=z*) --cdf(z*)
print('')
print('z=', z),print('')
print('P(X < 162) = ', stats.norm.cdf(z))
```

## Ejemplo 2:

Si $X \sim  \mathcal{N}(\mu=10,\,\sigma^{2}=4)$, calcule $P(X>13)$.

```python

mu = 10
sigma_2 = 4
sd = np.sqrt(sigma_2)
z = (13-mu)/sd

# P(X>13) = 1 - P(X<=13) = 1- cdf(13)

print('')
print('z=', z),print('')
print('P(X > 13)=', 1-stats.norm.cdf(z))
```

## Ejemplo 3

## Ejemplo 4

El peso m√©dio de 500 estudiantes del sexo masculino de una determinada universidad es 70 Kg y la desviaci√≥n estandar es 5 Kg. Asumiendo que los pesos son normalmente distribuidos, de forma aproximada.

(a) Determine el porcentaje de estudiantes que pesan entre 65 Kg y 75 Kg.

$$
\begin{aligned}
P(65\leq X \leq 75) &=  P\left(\frac{65-\mu}{\sigma} \leq \frac{X-\mu}{\sigma} \leq \frac{75 - \mu}{\sigma} \right)  \\
& =  P\left(\frac{65-70}{5} \leq Z \leq \frac{75 -70}{5} \right)   \\
& =  P\left(-1 \leq Z \leq 1 \right)   \\
& = P(Z \leq 1) - P(Z \leq -1) = 0,68.
\end{aligned}
$$

```python
mu = 70
sd = 5
z1 = (65-mu)/sd
z2 = (75 -mu)/sd
print('')
print('z1=',z1),print('')
print('z2=',z2),print('')
print('P(-1<=Z<=1)',stats.norm.cdf(z2)-st.norm.cdf(z1))
```

```python
mu = 70
sd = 5
n = 10000
X = np.random.normal(mu, sd, n)
m = 0
for x in X:
    if x > 65 and x < 75:
        m = m + 1
print('')
print('P(65<Z<75) (simulaci√≥n):', m/n)
```

(b) Determine el porcentaje de estudiantes que pesan m√°s de 80 Kg

$$

P(X > 80) &=  P\left(\frac{X-\mu}{\sigma} > \frac{80 - \mu}{\sigma} \right)  \\
& =  P\left(Z > \frac{80 -70}{5} \right)\\
& =  P\left( Z > 2 \right)   \\
& = 1 - P(Z \leq 2) = 0.02275.
\end{aligned}
$$

```python
z1 = (80-mu)/sd
p = 1-st.norm.cdf(z1)
print('P(X>80):',p)
```

## Ejemplo 5

Una maquina de bebidas est√° regulada de modo que sirve una media de 150ml por vaso. Si la cantidad servida por vaso sisgue una distribuci√≥n normal con desviaci√≥n estandar de 20 ml. Determine el porcentaje de vasos que tendr√°n mas de 175ml de bebida.

$$
P(X \geq 175) = P\left( \frac{X-\mu}{\sigma}\geq \frac{175-\mu}{\sigma}\right) = P\left( Z \geq \frac{175-\mu}{\sigma}\right) = 1 - P\left( Z < \frac{175-\mu}{\sigma}\right)
$$

```python
mu = 150
sd = 20
z = (175-mu)/sd
print('')
print((1-stats.norm.cdf(z))*100,'%')
```

Comparando la curva te√≥rica con simulaci√≥n

```python
def normal_dist(x , mean , sigma):
    prob_density = (1/(sigma*(math.sqrt(2*np.pi))))*np.exp(-0.5*((x-mean)/sigma)**2)
    return prob_density

mean = 75
sigma = 20
n = 10000
# Distribuci√≥n por simulaci√≥n
X = np.random.normal(mean, sigma, n)
plt.figure(figsize=(20,3))
Pk, bins, ignored = plt.hist(X, bins='auto', density=True, color='#0504aa',alpha=0.7,
                                rwidth=0.9)


x = np.linspace(np.min(X),np.max(X),200)
# Distribuci√≥n te√≥rica
pdf = normal_dist(x,mean,sigma)

plt.plot(x,pdf , color = 'red')
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(X)', fontsize = 16)
plt.show()
```

## Ejemplo 6

Una f√°brica usa botellas de gaseosa de 260 ml.  Una maquina es usada para llenas esas botellas. Asumiendo que la cantidad colocada por la maquina sigue una distribuci√≥n normal con media 250 ml y desviaci√≥n estandar de 10 ml. ¬øCual es el porcentaje de botellas que se derramar√°n?

```python
# P(X>260) = 1-P(X<=260)

mu = 250
sd = 10
x = 260
z = (x-mu)/sd
print((1-stats.norm.cdf(z))*100,'%')
```

## Ejemplo 7

 En una ciudad, 46\% de la poblaci√≥n vota en cierto candidato para alcalde. Una muestra de 500 personas es sorteada. Usando la aproximaci√≥n de la distribuci√≥n de la binomial por la normal. Calcule la probabilidad de que al menos 250 personas votar√°n por ese candidato.


```python
# P(X>250)
# p=0.46 (√©xito)

n = 500
p = 0.46
mu = n*p
sd = np.sqrt(n*(1-p)*p)
x = 250
z = (x-mu)/sd
print((1-stats.norm.cdf(z)))
```
